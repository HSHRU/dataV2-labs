# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here:
salaries.dtypes
```

```python
#salaries['Hourly Rate'] = (salaries['Hourly Rate']).fillna('')
```

```python
salaries1 = salaries[salaries['Hourly Rate'].notna()]
salaries1['Hourly Rate'] = salaries1['Hourly Rate'].str.replace('$', '', regex=True)
salaries1['Hourly Rate'] = pd.to_numeric(salaries1['Hourly Rate'])
```

```python


#salaries['Hourly Rate'] = salaries['Hourly Rate'].str.replace('$','', regex=True)
#salaries['Hourly Rate'] = pd.to_numeric(salaries['Hourly Rate'])
```

```python
salaries['Hourly Rate']
```

```python
# we can find a confidence interval for the population based on our sample
import scipy.stats

confidence_level = 0.95
degrees_freedom = len(salaries1['Hourly Rate']) - 1  # or, for large values of sample size, just use that
sample_mean = np.mean(salaries1['Hourly Rate'])
# note that we use the standard error of the sample 
# as an estimate of the standard error of the population (which is used in the theoretical formula)
sample_standard_error = scipy.stats.sem(salaries1['Hourly Rate']) # sem = standard error of the mean = std(mean)/sqrt(samplesize)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```

```python
from scipy.stats import ttest_1samp

stat, pval = ttest_1samp(salaries1['Hourly Rate'], 30.00)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)

```

```python
# waht does this mean is
# pvalue = probability of getting the result (mean) that we have, 
#          IF the poplulation mean is equal to 30

# p is low, null must go: we reject the H0
# so we assume the actual population mean is different from 30


# stat gives the distance (in sample sem (standard error of the mean)) of the hypothesised value to the sample mean
```

# Bonus Challenge - Hypothesis Tests of Proportions

Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction. 

You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).

In the cell below, use the `proportions_ztest` function from `statsmodels` to perform a hypothesis test that will determine whether the number of hourly workers in the City of Chicago is significantly different from 25% at the 95% confidence level.

```python
salaries4 = salaries[salaries['Salary or Hourly']=='Hourly']
salaries4['Salary or Hourly']=  np.where(salaries4['Salary or Hourly']=='Hourly', 1, 0)


#data_4 = salaries
#data_4['Salary or Hourly']=  np.where(data_4['Salary or Hourly']=='Hourly', 1, 0)
```

```python
from statsmodels.stats.proportion import proportions_ztest

count = 0.25*(len(salaries4['Salary or Hourly'] == 1)-1)
nobs = len(salaries4['Salary or Hourly'] == 1)-1
value =0.05
stat, pval = proportions_ztest(count, nobs, value)
print('{0:0.3f}'.format(pval))
```
